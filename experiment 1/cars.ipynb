```python
data = data.drop(['S.No.'], axis=1)
data.info()
```

```python
from datetime import date
date.today().year
data['Car_Age']=date.today().year-data['Year']
data.head()
```

```python
data['Brand']= data.Name.str.split().str.get(0)
data['model']= data.Name.str.split().str.get(1)+data.Name.str.split().str.get(2)
data['variant']=data.Name.str.split().str.get(3)
data[['Name', 'Brand', 'model', 'variant']]
```

```python
print(data.Brand.nunique())
```

```python
searchfor = ['Isuzu','ISUZU','Land','Mini']
data[data.Brand.str.contains('|'.join(searchfor))].head(5)
```

```python
data['Brand'].replace({'ISUZU':'Isuzu',"Mini": "Mini Cooper","Land":"Land Rover"}, inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
brand_count = data['Brand'].value_counts()
plt.figure(figsize=(12, 6))
sns.barplot(x=brand_count.index, y=brand_count.values)
plt.title('Count of Cars by Brand')
plt.xlabel('Brand')
plt.ylabel('Number of cars')
plt.xticks(rotation=90)
plt.show()
```

```python

brand_count = data['Brand'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(brand_count.values, labels=brand_count.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Cars by Brand')
plt.axis('equal')  # Ensures pie chart is a circle
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data['Car_Age'], bins=15, color='skyblue', edgecolor='black')
plt.title('Distribution of Car Age')
plt.xlabel('Car Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Car_Age'], data['Brand'], color='blue', alpha=0.6)
plt.title('Car Age vs Selling Price')
plt.xlabel('Car Age')
plt.ylabel('Selling Price')
plt.show()
```
